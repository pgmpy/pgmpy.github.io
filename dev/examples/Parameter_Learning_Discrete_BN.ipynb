# Parameter Learning in Discrete Bayesian Networks

In this notebook, we demonstrate examples of learning the parameters (CPDs) of a Discrete Bayesian Network given the data and the model structure. pgmpy has three main algorithms for learning model parameters:

1. **Maximum Likelihood Estimator** (`pgmpy.parameter_estimator.DiscreteMLE`): Simply estimates the Maximum Likelihood estimates of the parameters.
2. **Bayesian Estimator** (`pgmpy.parameter_estimator.DiscreteBayesianEstimator`): Allows users to specify priors.
3. **Expectation Maximization** (`pgmpy.parameter_estimator.DiscreteEM`): Enables learning model parameters when latent variables are present in the model.

Each of the parameter estimators follows the same usage pattern: initialize the estimator, call `fit(model, data)`, and access the learned CPDs via the `parameters_` attribute.

## Step 0: Generate some simulated data and a model structure

To do parameter estimation we require two things:
1. **Data**: For the examples, we simulate some data from the alarm model (https://www.bnlearn.com/bnrepository/discrete-medium.html#alarm) and use it to learn back the model parameters.
2. **Model structure**: We also need to specify the model structure to which to fit the data to. In this example, we simply use the structure to the alarm model.

```python
from pgmpy.example_models import load_model
from pgmpy.models import DiscreteBayesianNetwork

# Load the alarm model and simulate data from it.
alarm_model = load_model("bnlearn/alarm")
samples = alarm_model.simulate(n_samples=int(1e3))

print(samples.head())

# Define a new model with the same structure as the alarm model.
new_model = DiscreteBayesianNetwork(ebunch=alarm_model.edges())
```

## Using the Maximumum Likelihood Estimator

```python
from pgmpy.parameter_estimator import DiscreteMLE

# Initialize the estimator and fit it on the model structure and the data.
mle_est = DiscreteMLE()
mle_est.fit(new_model, samples)
```

```python
# The learned CPDs are stored in `parameters_`. Pick the one for the node `FIO2`.
fio2_cpd = next(cpd for cpd in mle_est.parameters_ if cpd.variable == "FIO2")
print(fio2_cpd)
```

```python
# The CPD for node `CVP`.
next(cpd for cpd in mle_est.parameters_ if cpd.variable == "CVP")
```

```python
# Add the estimated CPDs to the model.
new_model.add_cpds(*mle_est.parameters_)

# Check if the CPDs are added to the model.
new_model.get_cpds('PCWP')
```

## Using the Bayesian Estimator

```python
from pgmpy.parameter_estimator import DiscreteBayesianEstimator
```

`DiscreteBayesianEstimator` allows a few different ways to specify the priors. The prior type is chosen via the `prior_type` constructor argument:

1. **Dirichlet prior** (`prior_type="dirichlet"`): Requires specifying `pseudo_counts`. The `pseudo_counts` argument specifies the priors used for CPD estimation.
2. **BDeu prior** (`prior_type="BDeu"`): Requires specifying `equivalent_sample_size`, which is used to compute the priors.
3. **K2** (`prior_type="K2"`): Shorthand for a Dirichlet prior with `pseudo_counts=1`.

```python
# Pick the CPD for `FIO2` learned with a BDeu prior.
bdeu_est = DiscreteBayesianEstimator(prior_type="BDeu", equivalent_sample_size=1000)
bdeu_est.fit(new_model, samples)
print(next(cpd for cpd in bdeu_est.parameters_ if cpd.variable == "FIO2"))

# Pick the CPD for `CVP` learned with a Dirichlet prior.
dirichlet_est = DiscreteBayesianEstimator(prior_type="dirichlet", pseudo_counts=100)
dirichlet_est.fit(new_model, samples)
print(next(cpd for cpd in dirichlet_est.parameters_ if cpd.variable == "CVP"))
```

```python
k2_est = DiscreteBayesianEstimator(prior_type="K2")
k2_est.fit(new_model, samples)
k2_est.parameters_
```

## Using Expectation Maximization 

The Expectation Maximization (EM) estimator can work in the case when latent variables are present in the model. To simulate this scenario, we will specify some of the variables in our `new_model` as latents and drop those variables from `samples` to simulate missing data.

```python
model_latent = DiscreteBayesianNetwork(alarm_model.edges(), latents={'HISTORY', 'CVP'})
samples_latent = samples.drop(['HISTORY', 'CVP'], axis=1)
```

```python
from pgmpy.parameter_estimator import DiscreteEM

em_est = DiscreteEM()
em_est.fit(model_latent, samples_latent)
em_est.parameters_
```

## Shortcut for learning and adding CPDs to the model

The `BayesianNetwork` class also provies a `fit` method that acts as a shortcut way to estimate and add CPDs to the model.

```python
# Shortcut for learning all the parameters and adding the CPDs to the model.

model_struct = DiscreteBayesianNetwork(ebunch=alarm_model.edges())
model_struct.fit(data=samples, estimator=DiscreteMLE())
print(model_struct.get_cpds("FIO2"))

model_struct = DiscreteBayesianNetwork(ebunch=alarm_model.edges())
model_struct.fit(
    data=samples,
    estimator=DiscreteBayesianEstimator(prior_type="BDeu", equivalent_sample_size=1000),
)
print(model_struct.get_cpds("FIO2"))
```
